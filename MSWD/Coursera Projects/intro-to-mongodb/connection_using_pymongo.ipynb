```python
import pymongo

from pymongo import MongoClient

import pprint

from IPython.display import clear_output

client =MongoClient("mongodb+srv://rk:190031187@first-cpozf.mongodb.net/trial?retryWrites=true&w=majority")

pipeline = [
    {
        '$sortByCount':"$language"
    },
    {
        '$facet':{
            'top language combinations':[{'$limit':100}],
            'unusual combination shared by':[{
                '$skip':100
            },
            {
                '$bucketAuto':{
                    'groupBy':"$count",
                    'buckets':10,
                    'output':{
                        'language combinations':{'$sum':1}
                    }
            }
            }]
        }
    }
]

clear_output()

pprint.pprint(list(client.trial.first.aggregate(pipeline)))
```
